```python
from db import getConn
from datetime import datetime, timedelta 
import random 
import calendar

class GeraViagem():

turistas = [] 
idviagem = 0

def __init__(self): 
self.get_turistas() 
self.idviagem = self.get_id_viagem()

def get_turistas(self): 
con = getConn() 
cursor = con.cursor() 
sql = 'select * from turista' cursor.execute(sql) 
self.turistas = cursor.fetchall() 
con.close()

def get_turista_ramdom(self): 
return random.choice(self.turistas)

def get_resumos(self,): 
con = getConn() 
cursor = con.cursor() 
sql = 'select * from turistas where ano between 2010 and 2015 and ordemcontinente <> 998 and chegadas > 0' 
cursor.execute(sql) 
turistas = cursor.fetchall() 
con.close() 
return turistas

def get_data_chegada(self, mes, ano): 
def random_date(start, end): 
""" This function will return a random datetime between two datetime objects. """ 
delta = end - start 
int_delta = (delta.days * 24 * 60 * 60) + delta.seconds 
random_second = random.randrange(int_delta) 
return (start + timedelta(seconds=random_second)).strftime('%Y-%m-%d') 

a = datetime.strptime(str(mes) + '/01/' + str(ano) + ' 1:30 PM', '%m/%d/%Y %I:%M %p')
d = calendar.monthrange(ano, mes)[1] 
b = datetime.strptime(str(mes) + '/' + str(d) + '/' + str(ano) + ' 4:50 AM', '%m/%d/%Y %I:%M %p') 
return random_date(a, b)

def get_id_viagem(self): 
con = getConn() 
cursor = con.cursor() 
sql = 'select max(id_viagem) from viagem' 
cursor.execute(sql) 
r = cursor.fetchall() 
v = 0 
for t in r: 
v = t[0] 
if not v: 
v = 0 
con.close() 
return v + 1

def get_via(self): 
a = random.randrange(1, 5) 
if a == 1: 
return "Aérea" 
elif a == 2: 
return "Fluvial" 
elif a == 3: 
return "Marítima" 
else: 
return "Terrestre"

def generate_viagems(self, resumo): 
a = resumo[11] 
i = 1 
sql = "insert into viagem (data_chegada, pais_origem, continente_origem, num_passaporte, cod_uf, id_viagem, via)" 
sql += "values ('{}', '{}', '{}', '{}', {}, {}, '{}')" 
con = getConn() 
cursor = con.cursor()
while i <= a: 
turis = self.get_turista_ramdom() cursor.execute(sql.format(self.get_data_chegada(resumo[10], resumo[8]), resumo[2], resumo[0], turis[0], resumo[5], self.idviagem, self.get_via())) 
self.idviagem += 1 
i += 1
con.commit() 
con.close()

def generate(): 
gera = GeraViagem() 
resumos = gera.get_resumos() 
for res in resumos:
gera.generate_viagems(res) 
print('Inserido viagens.')
```
